```python
df = pd.concat([df_c, df_e])
df.head()
```

```python
import numpy as np
np.random.seed(7)
import sklearn.utils

# Add row id
df['row_id'] = df.index

# Create a Day of Week feature
df['DOW'] = df['Date'].str.slice(start=0, stop=3)

# Remove missing data
df.dropna(inplace=True)

# Add a binary column Experiment to denote
# if the data was part of the experiment or not (Random)
df['Experiment'] = np.random.randint(2, size=len(df))

# Remove missing data
df.dropna(inplace=True)

# Remove Date and Payments columns
del df['Date'], df['Payments']

# Shuffle the data
df = sklearn.utils.shuffle(df)
```

```python
df.head()
```

```python
# Reorder the columns
df = df[['row_id', 'Experiment', 'Pageviews', 'Clicks', 'DOW', 'Enrollments']]
```

```python
# prompt: LABLE ENCODING

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['DOW'] = le.fit_transform(df['DOW'])
df.head()
```

```python
X = df.loc[:, df.columns != 'Enrollments']
y = df['Enrollments']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
maxR2_Score=0
maxRS=0
for i in range(1,1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.30,random_state=i)
    LR = LinearRegression()
    LR.fit(X_train, y_train)
    predrf = LR.predict(X_test)
    score = r2_score(y_test, predrf)
    if score>maxR2_Score:
        maxR2_Score=score
        maxRS=i
print("Best accuracy is",maxR2_Score*100," on Random_state ",maxRS)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=maxRS)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,BaggingRegressor,ExtraTreesRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor,ExtraTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
import plotly.graph_objects as go
```

```python
model=[LinearRegression(),RandomForestRegressor(),DecisionTreeRegressor(),KNeighborsRegressor(),GradientBoostingRegressor(),Ridge(),SVR(),ExtraTreeRegressor(),ExtraTreesRegressor(),AdaBoostRegressor(),BaggingRegressor(),Lasso(),XGBRegressor()]
for m in model:
    print("*******************************************************************************")
    print("===============================================================================")
    print('accuracy score of :', m)
    m.fit(X_train,y_train)
    pred = m.predict(X_test)
    print("R2 Score:                ", r2_score(y_test,pred)*100)
    print("Mean Absolute Error:     ", mean_absolute_error(y_test,pred)*100)
    print("Mean Squared error:      ", mean_squared_error(y_test,pred)*100)
    print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(y_test,pred)*100))
    score=cross_val_score(m,X,y,cv=8)
    print(score)
    print("cross validation score:  ",score.mean())
    print("Difference between R2 score and cross validatio score is :  ",r2_score(y_test,pred)-abs(score.mean()*100))
    print('\n')
    # Gauge meter
    metric_value = r2_score(y_test,pred)*100
    fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=metric_value,
    title={'text': "Model Accuracy (%)"},
     gauge={'axis': {'range': [0, 100]},
        'bar': {'color': "darkblue"},
        'steps': [
            {'range': [0, 50], 'color': "red"},
            {'range': [50, 80], 'color': "yellow"},
            {'range': [80, 100], 'color': "green"}],}))

    # Display the gauge
    fig.show()
    print("===============================================================================")
    print("*******************************************************************************")
```

```python
# prompt: give me top 5 models

models_scores = []
for m in model:
    m.fit(X_train, y_train)
    pred = m.predict(X_test)
    r2 = r2_score(y_test, pred)*100
    models_scores.append((m.__class__.__name__, r2))

# Sort models by R2 score in descending order
models_scores.sort(key=lambda x: x[1], reverse=True)

# Print top 5 models
print("Top 5 Models:")
for i in range(min(5, len(models_scores))):
    print(f"{i+1}. {models_scores[i][0]} - R2 Score: {models_scores[i][1]}")
```

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
X_train['Enrollments'] = y_train
X_test['Enrollments'] = y_test
```

```python
X_train_h2o = h2o.H2OFrame(X_train)
X_test_h2o = h2o.H2OFrame(X_test)
```

```python
features = X_train.columns.values.tolist()
target = "Enrollments"
```

```python
# Construct the AutoML pipeline
auto_h2o = H2OAutoML()
# Train
auto_h2o.train(x=features,
               y=target,
               training_frame=X_train_h2o)
```

```python
# Overview of how the models performed
auto_h2o.leaderboard
```

```python
# Extract the best model from the leaderboard
auto_h2o = auto_h2o.leader
```

```python
X_test_temp = X_test.copy()
del X_test_temp['Enrollments']
```

```python
# Employing the model to make inference
X_test_h2o_copy = h2o.H2OFrame(X_test_temp)
y_preds = auto_h2o.predict(X_test_h2o_copy)

# Convert the predictions to a native list
y_preds = h2o.as_list(y_preds["predict"])
```

```python
def calculate_metrics(y_true, y_pred):
  print("R2 Score:", r2_score(y_true, y_pred)*100)
  print("Mean Absolute Error:", mean_absolute_error(y_true, y_pred))
  print("Mean Squared Error:", mean_squared_error(y_true, y_pred))
  print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_true, y_pred)))
```

```python
calculate_metrics(y_test, y_preds)
```
